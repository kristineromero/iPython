#### Ada - Fitting Stochastic Boosting Model ####
+ Gradient Boosting - a special case of the functional gradient descent view of boosting
+ Only for binary response
<br>

<font size = 1><a href="http://en.wikipedia.org/wiki/Gradient_boosting#Stochastic_gradient_boosting">source</a></font>

```python
%%R
library(ada)
data(iris)
```

```python
%%R
iris <- iris[iris$Species!="setosa",]
n <-dim(iris)[1]  # Total number of observations
trind<-sample(1:n,floor(.6*n),FALSE) # 60% random sample of observations
teind<-setdiff(1:n,trind) # 40% of observations not part of sample
iris[,5]<- as.factor((levels(iris[,5])[2:3])[as.numeric(iris[,5])-1]) # reset factor levels to not include setosa
```

```python
%%R
gdis<-ada(Species~.,data=iris[trind,],iter=20,nu=1,type="discrete") # fit 8-split trees
gdis=addtest(gdis,iris[teind,-5],iris[teind,5])
plot(gdis,TRUE,TRUE) # plot gdis
varplot(gdis) #variable selection plot
pairs(gdis,iris[trind,-5],maxvar=2) #pairwise plot
```

```python
%%R
perf = data.frame('predict' = predict(gdis, iris[teind,]),
                  'actual' = iris[teind,5])
table(perf)
# 86% success rate on iris
```

```python
%%R 
## rpart library should be loaded
data(iris)
iris.adaboost <- boosting(Species~., data=iris, boos=TRUE,
mfinal=10)
importanceplot(iris.adaboost)
sub <- c(sample(1:50, 25), sample(51:100, 25), sample(101:150, 25))
iris.bagging <- bagging(Species ~ ., data=iris[sub,], mfinal=10)
iris.predbagging<- predict.bagging(iris.bagging, newdata=iris[-sub,])
iris.predbagging$confusion
iris.predbagging$error
```

### Boosting vs Bagging ###
<font size = 1><a href="http://www.mathworks.com/help/stats/ensemble-methods.html#btgamju">source</a></font>

+ <b> Bag </b> generally constructs deep trees. This construction is both time consuming and memory-intensive. This also leads to relatively slow predictions.
+  <b> Boost </b> algorithms generally use very shallow trees. This construction uses relatively little time or memory. However, for effective predictions, boosted trees might need more ensemble members than bagged trees. Therefore it is not always clear which class of algorithms is superior.
+  <b> Bag </b>can estimate the generalization error without additional cross validation. See oobLoss.
+ Except for Subspace,<b> all boosting and bagging algorithms </b> are based on tree learners. Subspace can use either discriminant analysis or k-nearest neighbor learners.
